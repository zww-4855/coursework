```python
import numpy as np
from numpy import linalg as LA
```

```python
def calcNorm(p,Amatrix):
    if p==100: # call this infinity Norm
        maxSum=0.0
        for i in range(4):
            total=0.0
            for j in range(4):
                total=total+abs(Amatrix[i][j])
                if total>maxSum:
                    maxSum=total
        
        return maxSum
    else:
        total=0.0
        for i in range(10):
            total=total+abs(Amatrix[i])**p
        return total**(1.0/p)
    
```
