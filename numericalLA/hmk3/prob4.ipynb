```python
import numpy as np
from numpy import linalg as LA
import time
from datetime import datetime
```

```python
def calcNorm(p,Amatrix,rowDim,colDim):
    if p==100: # call this infinity Norm
        maxSum=0.0
        for i in range(rowDim):
            total=0.0
            for j in range(colDim):
                total=total+abs(Amatrix[i][j])
                if total>maxSum:
                    maxSum=total
        
        return maxSum
    else: # ANY OTHER VECTOR P NORM
        total=0.0
        for i in range(10):
            total=total+abs(Amatrix[i])**p
        return total**(1.0/p)
    
```
